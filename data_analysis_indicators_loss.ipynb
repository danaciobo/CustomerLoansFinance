```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('processed_loan_data.csv')

# selecting charged off & behind on payment customers
charged_off = df[df['loan_status'] == 'Charged Off']
behind = df[df['loan_status'].isin(['Late (16-30 days)', 'Late (31-120 days)', 'In Grace Period'])]


# bar chart to visualize if the loan grade is a risk factor
plt.figure(figsize=(10, 5))

sns.countplot(data=charged_off, x='grade', order=charged_off['grade'].value_counts().index, palette='viridis')
plt.title('Loan Grades - Charged Off')
plt.show()

sns.countplot(data=behind, x='grade', order=behind['grade'].value_counts().index, palette='rocket')
plt.title('Loan Grades - Behind on Payment')
plt.show()

# bar chart to visualize if purpose of the loan is a risk factor
plt.figure(figsize=(10, 5))

sns.countplot(data=charged_off, x='purpose', order=charged_off['purpose'].value_counts().index, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Loan Purpose - Charged Off')
plt.show()

sns.countplot(data=behind, x='purpose', order=behind['purpose'].value_counts().index, palette='rocket')
plt.xticks(rotation=45, ha='right')
plt.title('Loan Purpose - Behind on Payment')
plt.show()

# bar chart to visualize if  home ownership is a risk factor
plt.figure(figsize=(10, 5))
sns.countplot(data=charged_off, x='home_ownership', palette='viridis')
plt.title('Home Ownership - Charged Off')
plt.show()

sns.countplot(data=behind, x='home_ownership', palette='rocket')
plt.title('Home Ownership - Behind on Payment')

# bar chart to visualize if employment length is a risk factor
plt.figure(figsize=(10, 5))
sns.countplot(data=charged_off, x='employment_term_years', order=charged_off['employment_term_years'].value_counts().index, palette='viridis')
plt.title('Employment Length Years - Charged Off')
plt.show()

sns.countplot(data=behind, x='employment_term_years', order=behind['employment_term_years'].value_counts().index, palette='rocket')
plt.title('Employment Length Years - Behind on Payment')

# bar chart to visualize if verification status is a risk factor
plt.figure(figsize=(10, 5))
sns.countplot(data=charged_off, x='verification_status', order=charged_off['verification_status'].value_counts().index, palette='viridis')
plt.title('Verification Stetus - Charged Off')
plt.show()

sns.countplot(data=behind, x='verification_status', order=behind['verification_status'].value_counts().index, palette='rocket')
plt.title('Verification status - Behind on Payment')
plt.show()

print("""
Key Insights from the Analysis:
 - Loan Grades: Customers with middle grades, B, C and D are more likely to be charged off or fall behind on payments.
 - Loan Purpose: Certain purposes, such as 'credit_card' and 'debt_consolidation', show a higher tendency for loans being charged off or falling behind on payments.
 - Home Ownership: Renters seem to have a higher probability of falling behind or defaulting on loans compared to homeowners. But not by much.
 - Employment Length: Long employment terms (> 10 years) are more common among customers who have defaulted or are at risk. Also at 6 year mark.
 - Verification Status: I don't think there is a correlation here as more loans are likely to be verified so the figures for verified seem higher.
""")


```
