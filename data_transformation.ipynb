```python
from data_transform import DataTransform
from data_frame_transform import DataFrameTransform
from plotter import Plotter
from table_data import load_data_from_csv
from data_frame_info import DataFrameInfo
import pandas as pd
import numpy as np

# loads data from csv
file_path = "../loan_payments.csv"
df = load_data_from_csv(file_path)

# displays all columns
pd.set_option("display.max_columns", None)

# initializes DataFrameInfo for column summaries
data_frame_info = DataFrameInfo(df)

# prints summaries
data_frame_info.print_head()
print(data_frame_info.describe_columns())
print(data_frame_info.extract_statistics())
print(data_frame_info.count_distinct_values())
print(data_frame_info.get_shape())
print(data_frame_info.count_null_values())
print(data_frame_info.percentage_null_values())

# initializes DataTransform for initial processing
transformer = DataTransform(df)

# converts date columns to datetime format
date_columns = ["issue_date", "last_payment_date", "next_payment_date", "last_credit_pull_date", "earliest_credit_line"]
df = transformer.convert_to_datetime(date_columns)

# changes the 'employment_length' column to numeric
df = transformer.clean_employment_length("employment_length")

# converts 'term' to numeric and changes to term_months
df = transformer.convert_term_to_numeric("term")

# converts 'int_rate' and 'loan_amount' columns to numeric - just in case
df["int_rate"] = pd.to_numeric(df["int_rate"], errors="coerce")
df["loan_amount"] = df["loan_amount"].astype(float)

# converts categorical columns to category type
categorical_columns = ["grade", "sub_grade", "home_ownership", "verification_status", "loan_status", "payment_plan", "purpose", "application_type"]
df = transformer.convert_categorical_columns(categorical_columns)

# initializes DataFrameInfo to see column summaries after transformations
df_info = DataFrameInfo(df)

# prints updated column summaries
df_info.print_head()
print(df_info.describe_columns())
print(df_info.extract_statistics())
print(df_info.count_distinct_values())
print(df_info.get_shape())
print(df_info.count_null_values())
print(df_info.percentage_null_values())


# initializes Plotter and DataFrameTransform for more complex transformations
plotter = Plotter(df)
transformer = DataFrameTransform(df)

# checks and plots null values before handling
print("Null values before handling:")
print(transformer.check_nulls())

plotter.plot_nulls(title="Missing Values Before Handling")

# drops columns with >50% missing data
df = transformer.drop_columns_by_nulls(threshold=50)

# checks null values after dropping columns
print("Null values after dropping columns with >50% missing data:")
print(transformer.check_nulls())

# applies calculated imputations for some fields
df = transformer.impute_calculated_values()

# checks null values after calculated imputations
print("Null values after calculated imputations:")
print(transformer.check_nulls())

# applies general imputation for remaining missing values
df = transformer.impute_missing_values()

# checks null values after complete handling - should be none
print("Null values after all imputations:")
print(transformer.check_nulls())

# reinitializes Plotter with updated DataFrame
plotter = Plotter(df)

# visualizes NULLs after handling
plotter.plot_nulls(title="Missing Values After Handling")



# Saving the cleaned DataFrame to a CSV file for use in calculations later
df.to_csv("financial_loan_data.csv", index=False)
print("Data transformation for financial analysis complete. Saved to 'financial_loan_data.csv'")



# identifies and handles skewed data
skewed_columns = transformer.detect_skewed_columns()

# excluding id & member id columns that should not be transformed
columns_to_exclude = ['id', 'member_id']
filtered_skewed_columns = [col for col in skewed_columns if col not in columns_to_exclude]

# prints skewed columns and visualizes them
print(f"Skewed columns: {filtered_skewed_columns}")
plotter.plot_skewness(filtered_skewed_columns, title="Skewed Columns Before Transformation")

# tries to correct skeweness and changes the df
df = transformer.transform_skewed_columns(filtered_skewed_columns)

# checks and visualizes skewed columns after transformations
skewed_columns_after = transformer.detect_skewed_columns(threshold=1.0)
print(f"Skewed columns after transformation: {skewed_columns_after}")
print("Distributions after transformation:")
plotter.df = df
plotter.plot_skewness(skewed_columns_after, title="Skewed Columns After Transformation")

# choosing the columns and excluding some irrelevant one to check for outiliers

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
exclude_columns = ["delinq_2yrs", "recoveries", "total_rec_late_fee", "collection_recovery_fee", "id", "member_id"]
numeric_columns = [col for col in numeric_columns if col not in exclude_columns]

# visualizing the outliers and printing total number of rows
plotter.plot_outliers(numeric_columns)

print(f"Before outlier removal: {df.shape[0]} rows")


# removing outliers from the data

df = transformer.remove_outliers(numeric_columns)

# Re-visualize the data after removing outliers and printing number of rows left
print(f"After outlier removal: {df.shape[0]} rows")
print("Visualizing data after removing outliers:")
plotter.plot_outliers(numeric_columns, title="Boxplots of Numeric Columns (After Removing Outliers)")


# visualizing the correlation matrix before removing columns and printing number of columns
plotter.plot_correlation_matrix(title="Correlation Matrix Before Removing Highly Correlated Columns")
print(f"Shape Before Removing Columns: {df.shape}")

#removing highly correlated columns and print the removed columns
df = transformer.remove_highly_correlated_columns(threshold=0.9)

# checking the shape of the DataFrame after removing columns and revisualizing correlation matrix after transformation

print(f"Shape After Removing Columns: {df.shape}")

plotter.df = df
plotter.plot_correlation_matrix(title="Correlation Matrix After Removing Highly Correlated Columns")

# saving the processed data to a new csv file
df.to_csv("processed_loan_data.csv", index=False)
print("Data processing complete. Saved to 'processed_loan_data.csv'")



```
